```python
#Choose a numerical field and print its unique values
unique_values = data['DelayMinutes'].unique()
print(unique_values)
```

```python
#Replace an extreme set of values in the dataset with np.NaN

data.loc[data['DelayMinutes'] <0, 'DelayMinutes'] = np.NaN 
data.head(5) #NAN is a float value in pandas
```

```python
#Plot a histogram of a numerical variable

plt.figure(figsize = (8,4))
plt.hist(data['DelayMinutes'])

plt.title('Histogram of Delay Minutes', fontsize=12)
plt.xlabel('Delay Minutes', fontsize=8)
plt.ylabel('Frequency', fontsize=8)
plt.show()
```

```python
#Choose a categorical field and represent it as a numerical field, then display the last few rows


drop_columns = ['ScheduledDeparture','ActualDeparture','ScheduledArrival','ActualArrival','Diverted', 'TailNumber']
data = data.drop(columns = drop_columns)

categorical_column = 'Cancelled'
data[categorical_column + '_Numerical'] = pd.Categorical(data[categorical_column]).codes

data.tail()
```

```python
#Use the zscore function to standardize a numerical field, then show its histogram
from scipy.stats import zscore

df = zscore(data['Distance'])

plt.hist(df, bins = 200, edgecolor='black')
plt.title('Histogram of Standardized Distance')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#Identify a field and use a criterion of your choosing to filter for outliers 
#Create a new dataset with the outliers and display the first few rows

numerical_column = 'Distance'

data['Zscore'] = zscore(data[numerical_column])

outliers = data[(data['Zscore'] > 1) | (data['Zscore'] < -1)]

outliers.head()
```
